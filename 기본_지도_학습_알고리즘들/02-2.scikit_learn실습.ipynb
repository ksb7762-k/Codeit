## Multiple Linear Regression

```python
import pandas as pd
A = pd.read_csv('/content/drive/MyDrive/boston.csv')
```

```python
A.columns
```

```python
input = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']
```

```python
x = A[input]
x
```

```python
y = A[['MEDV']]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 5)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_ # 세타1~세타13
```

```python
model.intercept_   # 세타0
```

```python
y_test_prediction = model.predict(x_test)
y_test_prediction
```

```python
mean_squared_error(y_test, y_test_prediction)
```

```python
mean_squared_error(y_test, y_test_prediction) ** 0.5 # 이 모델로 가격을 예측하면 약 4500달러 정도 오차가 있다는 뜻
```
