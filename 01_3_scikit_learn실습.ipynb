```python
import pandas as pd
boston = pd.read_csv('/content/drive/MyDrive/boston.csv')
boston
```

```python
boston.describe()
```

```python
x = boston[['AGE']]
x
```

```python
y = boston[['MEDV']]
y
```

# scikit-learn 데이터 셋 나누기  

모델의 성능을 신빙성있게 평가하기 위해서 학습데이터와 평가데이터로 나눔

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=5)
```

test_size = 0.2 : 전체 데이터셋 중 20%만 테스트 셋으로 사용하고 나머지 80%는 트레이닝 셋으로 사용하라는 뜻   

random_state :  그 20%를 어떻게 고를지 정하는 파라미터 (optional parameter이기 때문에 안넘겨도 됨)   
안넘겨주면 실행할 떄마다 매번 랜덤한 20%를 테스트 셋으로 새롭게 고르게됨. 정수값을 넘겨주면 매번 20%의 똑같은 데이터를 고름   

train_test_split함수는 4개의 값을 튜플로 리턴하는데 그 4개의 값을 왼쪽 4개의 변수에 저장함!

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# scikit-learn으로 선형 회귀 쉽게 하기

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_ # theta_1의 값
```

```python
model.intercept_ # theta_0의 값
```

```python
# 우리가 구한 최적선 ==> f(x) = 31.04617413 - -0.12402883x
```

```python
y_test_prediction = model.predict(x_test)
y_test_prediction
```

 y_test에는 실제 목표변수의 값이 들어있음.  
 y_test_prediction과 비교해서 우리의 모델의 성능을 평가하려고 함   
 보틍은 RMSE를 사용함 ==> RMSE 구해주는 함수를 불러오자

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_test_prediction) ** 0.5 # 루트 씌우기
```
