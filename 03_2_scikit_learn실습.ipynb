# Polynomial Regression  

scikitlearn을 이용해서 다항회귀를 해보자  

```python
import pandas as pd
boston = pd.read_csv('/content/drive/MyDrive/boston.csv')
```

```python
boston
```

```python
input = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']
```

```python
x = boston[input]
```

```python
x.shape # 원래 타겟변수() 뺴야함 그래서  (206, 13)
```

```python
from sklearn.preprocessing import PolynomialFeatures # 다항속성을 만들어주는 툴
```

```python
polynomial_transformer = PolynomialFeatures(2) # 가설함수가 2차라고 가정
```

데이터를 다항회귀를 할 수 있게끔 가공

```python
polynomial_data = polynomial_transformer.fit_transform(x)
polynomial_data
```
