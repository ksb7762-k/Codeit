# 01. 같은 형식의 데이터 합치기 : concat()

```python
import pandas as pd

english_df1 = pd.DataFrame({
    'name' : ['dongwook', 'taeho', 'jimin'], 
    'english_score': [50, 86, 68]
})

english_df2 = pd.DataFrame({ 
    'name' : ['yoonsoo', 'sowon', 'haeun'],
    'english_score' : [88, 81, 72]
})
```

```python
english_df1
```

```python
english_df2
```

```python
pd.concat([english_df1, english_df2])
```

```python
pd.concat([english_df1, english_df2], ignore_index=True)
```

```python
english_df = pd.concat([english_df1, english_df2], ignore_index=True)
```

```python
english_df
```

```python
math_df = pd.DataFrame({
    'name' : ['dongwook', 'taeho', 'jimin', 'yoonsoo', 'sowon', 'haeun'], 
    'math_score': [86, 31, 91, 75, 68, 80]
})
```

```python
math_df 
```

```python
pd.concat([english_df, math_df])
```

```python
pd.concat([english_df, math_df], axis=1)
```

# 03. 조인(Join) 연산 : 공통으로 들어있는 값을 기준으로 데이터를 합치는 것

## Inner Join : 두 데이터에 모두 들어있는 값들만 찾아서 합침 
## Left Outer Join : 왼쪽 데이터를 기준으로 데이터를 합침 
## Right Outer Join : 오른쪽 데이터를 기준으로 데이터를 합침
## Full Outer Join : 두 데이터에 있는 값을 모두 가져와서 하나로 합침

# 08. merge()와 join() 더 알아보기 

### 지금까지 merge() 함수를 가지고 컬럼명을 기준으로 데이터를 합치는 방법과 join() 함수를 가지고 인덱스를 기준으로 데이터를 합치는 방법을 알아봤는데요. 반대로 merge()에서 인덱스를, join()에서 컬럼명을 사용해서 데이터를 합치는 것도 가능합니다.

# merge() 함수 더 알아보기

## 먼저 merge() 함수에서 인덱스를 키 값으로 사용하는 방법을 알아볼게요. 일단 영상에서 본 employee_df와 survey_df를 그대로 사용해서 설명드릴 거고요. 둘 다 사원 번호가 인덱스로 설정되어 있는 상태라고 가정하겠습니다.

```python
import pandas as pd

employee_df = pd.read_csv('data/employee.csv', index_col='id')
survey_df = pd.read_csv('data/survey.csv', index_col='id')
```

## merge()에서는 아래와 같이 left_index라는 값을 True로 설정하면 왼쪽 데이터에 있는 인덱스를 키 값으로 사용하고, right_index라는 값을 True로 설정하면 오른쪽 데이터에 있는 인덱스를 키 값으로 사용합니다.

```python
pd.merge(employee_df, survey_df, left_index=True, right_index=True)
```

## 참고로 둘 중 하나만 True로 설정하는 것도 가능합니다. 예를 들어 left_index만 True로 설정하면 왼쪽 데이터만 인덱스를 키 값으로 사용하고, 오른쪽 데이터는 특정 컬럼을 키 값으로 사용하게 되는데요. 이때, 반드시 right_on 옵션을 함께 사용해서 어떤 컬럼을 키 값으로 할 건지 정해 줘야 합니다. 반대로 right_index만 True로 설정할 땐, left_on 옵션을 함께 사용해야 하고요.

# join() 함수 더 알아보기
## join() 함수에서 컬럼을 키 값으로 사용하는 방법도 알아보겠습니다. 일단 아래와 같이 reset_index() 함수를 사용해서, employee_df의 사원 번호(id)를 다시 컬럼으로 바꿔 줄게요.

```python
employee_df = employee_df.reset_index()
```

## join()에서는 on이라는 옵션을 사용해서 왼쪽 데이터에서 키 값으로 사용할 컬럼을 지정할 수 있습니다. 예를 들어 아래 코드를 실행하면 employee_df 데이터의 id 컬럼과 survey 데이터의 인덱스를 기준으로 조인 연산을 수행합니다. 이때 주의할 점은 왼쪽 데이터만 인덱스와 컬럼 중에서 자유롭게 키 값을 선택할 수 있고, 오른쪽 데이터는 인덱스만 키 값으로 사용할 수 있다는 것입니다. 추가적으로, 지금 데이터에 공통적으로 department라는 컬럼이 있으니까, 오류가 나지 않게 rsuffix 값도 아래와 같이 설정해 줄게요.

```python
employee_df.join(survey_df, on='id', rsuffix='_x')
```

## 지금까지 알아본 것처럼, merge()와 join() 함수는 조인 연산에 사용된다는 공통점이 있지만 사용 방법에 조금씩 차이가 있습니다. 정리해 보자면 merge()는 보다 다양한 옵션을 사용할 수 있다는 장점이 있고, join()은 인덱스가 키 값으로 설정되어 있을 때 좀 더 간결한 코드로 데이터를 합칠 수 있다는 장점이 있다고 할 수 있습니다. 상황이나 기호에 따라 더 편하게 느껴지는 함수를 사용하면 되겠죠?
