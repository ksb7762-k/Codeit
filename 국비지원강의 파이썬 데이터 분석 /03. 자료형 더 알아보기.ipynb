# 6. 문자열 포매팅 

## 이번 레슨에서는 파이썬에서 문자열을 포함해서 값을 출력하는 방법을 몇 가지 소개해 드리겠습니다.

## 예를 들어 지금 아래 name, age 변수에 있는 값을 사용해서 ‘제 이름은 최지웅이고 32살입니다.’라는 문장을 출력한다고 할게요.

```python
name = "최지웅"
age = 32
```

## 지금 age가 정수형이니까, 이전에 배운 형 변환을 사용해서 코드를 작성해 줬는데요. 코드도 좀 길고 작성하기도 번거롭습니다. 문장이 복잡해지면 이런 문제가 더 심해질 수도 있죠.

```python
print("제 이름은 " + name + "이고 " + str(age) + "살입니다.")
```

## % 기호
## 이제는 잘 쓰지 않는 오래된 방식이기는 하지만, %s, %d와 같은 '포맷 스트링'이라는 것을 사용하는 방법도 있습니다. C나 자바 등 많은 언어들에서 이와 유사한 방식으로 문자열을 포매팅합니다.

```python

name = "최지웅"
age = 32

print("제 이름은 %s이고 %d살입니다." % (name, age))

```

## format() 메소드
## format()이라는 메소드를 사용하는 방법도 있어요. 아래와 같이 문자열 안에 값을 넣어줄 부분을 중괄호 {}로 대체해 주고, 뒤에 .format()이라고 쓴 다음에 함수 안에 중괄호 안에 들어갈 값들을 순서대로 파라미터로 넣어주면 됩니다.

```python

name = "최지웅"
age = 32

print("제 이름은 {}이고 {}살입니다.".format(name, age))

```

## f-string

```python

name = "최지웅"
age = 32

print(f"제 이름은 {name}이고 {age}살입니다.")

```

# 7. 불 대수(Boolean Algebra: 일상적인 논리를 수학적으로 표현한 것) 

## 1. 가능한 값이 True와 False 두 개 뿐 2. 연산이 AND, OR, NOT이 있음 

# 8. 불린형(불린으로 쓰려면 따옴표없이 써야함)

```python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

```python
print(not True)
print(not False)
```

```python
print(2 > 1)
print(2 < 1)
print(3 >= 2)
print(3 <= 3)
print(2 == 2)
print(2 != 2)
print('강세범' == '강세범')
print('강새범' != '강세범')
```

```python
print(2 > 1 and '강세범' == '강세범')
```

```python
print(2 > 1 or '강세범' == '강새범')
```

```python
print(not not True)
print(not not False)
```

```python
print(7 == 7 or (4 < 3 and 12 > 10))
```

```python
x = 3
print(x > 4 or not (x < 2 or x == 3))
```
